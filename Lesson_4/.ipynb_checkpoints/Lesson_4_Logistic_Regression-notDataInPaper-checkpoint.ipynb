## Defaulting on loans

Suppose we want to predict whether someone will default on their loan based on their income and the amount that they borrowed.

We will look at the dataset Default-Data which has three columns:  the first column is the loan amount, the second column is income, and the third column is the status of the loan.  

If the borrower has defaulted on their loan, we mark their status with a 1 in that column.  If the borrower has not defaulted on their loan, we mark their status as 0.

```python
#------------------------
# Import packages
#------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from numpy.linalg import norm

# Plot style
plt.style.use('seaborn-whitegrid')

#------------------------
# Make dataframe
#------------------------

# Creates a dataframe from file called Default-Data-New.csv
#---------------------
dfLoans = pd.read_csv('Default-Data-New.csv')

# Print some info about the dataset
#---------------------

# Prints first 5 lines of data
print('The first 5 lines of data look like:\n', dfLoans.head(5))

# Print basic stats on data
print('Some basic stats on our data are:\n', dfLoans.describe())
```

### Create plots
Let us plot our data with the x-axis displaying the borrower's loan amount, the y-axis the borrower's income, and the color of the data representing the borrower's status.  Recall that 0 means the borrower has not defaulted on their loan and 1 means that they have defaulted.

```python
#------------------------
# Create plot
#------------------------

# Make figure size
fig = plt.figure(dpi = 140)

# Plot df data
#---------------------
sns.scatterplot(data = dfLoans, x='loan_amount', y='income', hue='status', style = 'status')

# Label figure
plt.xlabel('Loan Amount (dollars)')
plt.ylabel('Income')
plt.title('Loan Status by Loan Amount and Income')

# Uncommment to save figure
#---------------------
#plt.savefig('loanStatus.pdf', bbox_inches='tight', dpi= 300)
 
# Show the plot
#---------------------
plt.show()
```

### Add logistic and linear regression models
We can use both logistic and linear regression here to see how they differ along with plotting the data.

```python
# With sklearn
#---------------------
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression

#------------------------
# Build logistic regression model using loan amount and income as predictors
#------------------------

# Split data into predictors x and output y
#---------------------

# Make list of predictors from dataset
predictors = ['loan_amount', 'income']

# Label predictors as vector x and output y
x = dfLoans[predictors]
y = dfLoans['status']

# Run logistic regression program
#---------------------
logRegModel = LogisticRegression(penalty = 'l2', solver = 'newton-cg').fit(x, y)

# Run linear regression program
#---------------------
linRegModel = LinearRegression().fit(x, y)
```

Now, we have our two models: the logistic model and the linear model to compare.  We can see our vector of theta's if we'd like.

```python
# theta0 is our intercept
#---------------------
theta0 = logRegModel.intercept_[0]

# theta consists of all other thetas
#---------------------
theta1 = logRegModel.coef_[0][0]
theta2 = logRegModel.coef_[0][1]

# Print out thetas
#---------------------
print('The thetas for our logistic regression model are:\ntheta0 =', theta0)
print('theta1 =', theta1)
print('theta2 =', theta2)
```

We can plot our new function if we'd like.  We start by creating some values for loan amounts ($x_1$) and incomes ($x_2$).  Then, we create our log-likelihood function.

```python
# Pick the number of points to plot
#---------------------
n = 200

# Values to plot for our function
#---------------------
x1 = np.linspace(min(dfLoans['loan_amount']), max(dfLoans['loan_amount']), n)
x2 = np.linspace(min(dfLoans['income']), max(dfLoans['income']), n)

# Plug our values into our function log-likelihood function
# 1/(1+exp(-(theta0 + theta1*loanAmount = theta2*income)))
#---------------------
logModelPlot = lambda x1, x2: 1/(1+np.exp(-theta0 -theta1*x1 - theta2*x2))
```

We can setup to do the same for our linear model to compare.

```python
# lintheta0 is our intercept
#---------------------
lintheta0 = linRegModel.intercept_

# lintheta consists of all other thetas
#---------------------
lintheta1 = linRegModel.coef_[0]
lintheta2 = linRegModel.coef_[1]

# Print out thetas
#---------------------
print('The thetas for our linear regression model are:\ntheta0 =', lintheta0)
print('theta1 =', lintheta1)
print('theta2 =', lintheta2)

# Plug our values into our function linear function
# 1/(1+exp(-(theta0 + theta1*loanAmount = theta2*income)))
#---------------------
linModelPlot = lambda x1, x2: lintheta0 + lintheta1*x1 + lintheta2*x2
```

### Plot data with our models
Now, we will plot our data and our two regression lines: 1) logistic regression, 2) linear regression.

```python
#-----------------------------
# Plot our data and functions 
#-----------------------------

# Make figure size
#---------------------
fig = plt.figure(dpi = 140)

# Plot data
#---------------------
sns.scatterplot(data = dfLoans, x = 'loan_amount', y = 'status', color='crimson', label = 'Data')
# Plot log model
plt.plot(x1, logModelPlot(x1,x2), '-', color = 'steelblue', linewidth = 2, label = 'Logistic Model')
# Plot linear model
plt.plot(x1, linModelPlot(x1,x2), '-', color = 'purple', linewidth = 2, label = 'Linear Model')

# Label axes + title
#---------------------
plt.xlabel('Loan Amount (dollars)')
plt.ylabel('Status')
plt.title('Loan Status')
plt.ylim(-0.1,1.1)

# Add legend
#---------------------
plt.legend()

# Uncommment to save figure
#---------------------
#plt.savefig('loanStatusLogLinCompare.pdf', bbox_inches='tight', dpi= 300)
 
# Show the plot
#---------------------
plt.show()
```

## Using Gradient Descent for Logistic Regression

In  general, if we have $N$ observations and two categories where $y^i$ is 0 or 1, we define the likelihood as

$$\begin{equation}
    \label{EdholmHohnRadunskayaeq:likelihood}
{\cal L}(\boldsymbol{\theta}) = 
\prod_{i|y^i = 1} p({\bf x}^i; \boldsymbol{\theta})
\prod_{i|y^i = 0} \left({1 - p({\bf x}^i; \boldsymbol{\theta})}\right)
\end{equation}$$

We can compactly write the log-likelihood, our objective function ${\ell}$, as:

$$\begin{eqnarray}
{\ell} (\boldsymbol{\theta})  &=& \sum_{i=1}^N \left[{ y^i \ln p({\bf x}^i; \boldsymbol{\theta}) + (1 - y^i) \ln(1 - p({\bf x}^i; \boldsymbol{\theta})}\right] \\
&=& \sum_{i=1}^N \left[{ y^i \boldsymbol{\theta}^T {\bf \tilde{x}}^i - \ln\left({1 + e^{\boldsymbol{\theta}^T {\bf \tilde{x}}^i } }\right) }\right],
\end{eqnarray}$$

where $P(Y = 1 | X = {\bf x}; \boldsymbol{\theta})= p({\bf x}^i; \boldsymbol{\theta}) =\dfrac{e^{\boldsymbol{\theta}^T\bf{\tilde{x}}}}{1 + e^{\boldsymbol{\theta}^T{\bf \tilde{x}}}}$

### Via gradient descent with the mean log likelihood function

```python
# Insert your modified code here
#----------------------------------------------------------------------
# Solutions for instructor are below - erase before giving to students
#----------------------------------------------------------------------
#------------------------
# Create functions
#------------------------

# Create the gradient vector
#------------------------
def gradf(X,y, Theta):
    # Create Xtilde = column of ones + X
    N = len(y)
    X0 = np.ones((N,1))
    Xtilde = np.column_stack((X0, X))
    # Compute scores
    scores = Xtilde@Theta # Nx1 vector
    # Compute probability function
    predictions = 1 / (1 + np.exp(-scores)) # Nx1
    # Compute gradient via dot product
    gradient = (y-predictions)@Xtilde # 1x3
    return gradient

#------------------------
# Create variables
#------------------------

from sklearn import preprocessing

# Name the data
#------------------------
x1 = dfLoans['loan_amount'].values
x2 = dfLoans['income'].values

# Normalize data
#------------------------
#normalized_x1 = preprocessing.normalize([x1])[0]
#normalized_x2 = preprocessing.normalize([x2])[0]
normalized_x1 = (x1-np.mean(x1))/np.std(x1)
normalized_x2 = (x2-np.mean(x2))/np.std(x2)

# Data to import into functions
#------------------------
X = np.column_stack((normalized_x1,normalized_x2))
y = dfLoans['status'].values


# Parameters
#------------------------
N = len(y)

# Starting point
Theta = [0,0,0]

# Step size to move along gradient
gamma = 0.01

# initialize num of steps to take
totalNumSteps = 0
# create a max number of steps to take
maxSteps = 200000
# how close we want to be to f
tolerance = 0.000001
# initial difference between old and new theta values
diff = 100

#------------------------
# Gradient descent calcs
#------------------------

while((totalNumSteps < maxSteps)&(diff > tolerance)):
    
    # Calculate new (x,y) values - move along gradient to new (x,y) position
    newTheta = Theta + 1/N*gamma*gradf(X,y,Theta)
    
    # Calculate the difference between current and new (x,y) values
    diff = norm(newTheta-Theta,2)
    
    # Reassign current (x,y) to new (x,y)
    Theta = newTheta
    
    # Add 1 to counter for total number of steps
    totalNumSteps += 1

# Get thetas
#------------------------
theta0 = Theta[0]
theta1 = Theta[1]
theta2 = Theta[2]

# Calculuate unscaled thetas
#------------------------
unscaled_theta0 = theta0 - theta1*np.mean(x1)/(np.std(x1)) - theta2*np.mean(x2)/(np.std(x2))
unscaled_theta1 = theta1/(np.std(x1))
unscaled_theta2 = theta2/(np.std(x2))

print('Our theta_0 value after', totalNumSteps, 'time steps is', theta0)
print('Our theta_1 value after', totalNumSteps, 'time steps is', theta1)
print('Our theta_2 value after', totalNumSteps, 'time steps is', theta2)

print('Our unscaled theta_0 value after', totalNumSteps, 'time steps is', unscaled_theta0)
print('Our unscaled theta_1 value after', totalNumSteps, 'time steps is', unscaled_theta1)
print('Our unscaled theta_2 value after', totalNumSteps, 'time steps is', unscaled_theta2)
```

### Example 3 - Logistic Regression and Gradient Descent  

Now, let us put the gradient to work. We will use gradient descent to implement logistic regression in order to solve a classification problem.

**The minimization problem**

We want to maximize the log likelihood function, so the objective function that we want to *minimize* is the negative of the log likelihood, whose partial derivatives we calculated in the equation above.

The update rule is 

$$
\boldsymbol{\theta}^{(t+1)}=\boldsymbol{\theta}^{(t)}-\gamma^{(t)}  \nabla (- \ell) =\boldsymbol{\theta}^{(t)}+\gamma^{(t)}  \nabla ( \ell)
$$

Here, we are writing $\boldsymbol{\theta}$, the gradient vector, $\nabla \ell$ and the input vector, ${\bf x}^{(i)}$, as *column* vectors. 

If we have a large dataset, we often want to look at the mean log likelihood function.  That is, we modify our log likelihood function by

$$
\boldsymbol{\theta}^{(t+1)}=\boldsymbol{\theta}^{(t)}+\frac{1}{N}\gamma^{(t)}  \nabla ( \ell)
$$
where $N$ is the number of data entries.

**Implementation**



```python
# Insert your modified code here
#----------------------------------------------------------------------
# Solutions for instructor are below - erase before giving to students
#----------------------------------------------------------------------
#------------------------
# Create functions
#------------------------

# Create the gradient vector
#------------------------
def gradf(X,y, Theta):
    # Create Xtilde = column of ones + X
    N = len(y)
    X0 = np.ones((N,1))
    Xtilde = np.column_stack((X0, X))
    # Compute scores
    scores = Xtilde@Theta # Nx1 vector
    # Compute probability function
    predictions = 1 / (1 + np.exp(-scores)) # Nx1
    # Compute gradient via dot product
    gradient = (y-predictions)@Xtilde # 1x3
    return gradient

#------------------------
# Create variables
#------------------------

# Name the data
#------------------------
x1 = dfAdmissions['gre'].values
x2 = dfAdmissions['gpa'].values

# Normalize data
#------------------------
normalized_x1 = (x1-np.mean(x1))/np.std(x1)
normalized_x2 = (x2-np.mean(x2))/np.std(x2)

# Data to import into functions
#------------------------
X = np.column_stack((normalized_x1,normalized_x2))
y = dfAdmissions['admit'].values


# Parameters
#------------------------
N = len(y)

# Starting point
Theta = [0,0,0]

# Step size to move along gradient
gamma = 0.01

# initialize num of steps to take
totalNumSteps = 0
# create a max number of steps to take
maxSteps = 200000
# how close we want to be to f
tolerance = 0.000001
# initial difference between old and new theta values
diff = 100

#------------------------
# Gradient descent calcs
#------------------------

while((totalNumSteps < maxSteps)&(diff > tolerance)):
    
    # Calculate new (x,y) values - move along gradient to new (x,y) position
    newTheta = Theta + 1/N*gamma*gradf(X,y,Theta)
    
    # Calculate the difference between current and new (x,y) values
    diff = norm(newTheta-Theta,2)
    
    # Reassign current (x,y) to new (x,y)
    Theta = newTheta
    
    # Add 1 to counter for total number of steps
    totalNumSteps += 1

# Get thetas
#------------------------
theta0 = Theta[0]
theta1 = Theta[1]
theta2 = Theta[2]

# Calculuate unscaled thetas
#------------------------
unscaled_theta0 = theta0 - theta1*np.mean(x1)/(np.std(x1)) - theta2*np.mean(x2)/(np.std(x2))
unscaled_theta1 = theta1/(np.std(x1))
unscaled_theta2 = theta2/(np.std(x2))

print('Our theta_0 value after', totalNumSteps, 'time steps is', theta0)
print('Our theta_1 value after', totalNumSteps, 'time steps is', theta1)
print('Our theta_2 value after', totalNumSteps, 'time steps is', theta2)

print('Our unscaled theta_0 value after', totalNumSteps, 'time steps is', unscaled_theta0)
print('Our unscaled theta_1 value after', totalNumSteps, 'time steps is', unscaled_theta1)
print('Our unscaled theta_2 value after', totalNumSteps, 'time steps is', unscaled_theta2)
```

### Exercise 4

Suppose we collect data for a group of students in a statistics class with variables $x_1 = $ hours studied and
$x_2 = $ undergrad GPA, and output $y = $ receive an A (1 or 0).  We fit a logistic regression and produce estimated coefficients $\theta_0 = -6$, $\theta_1 = 0.05$, $\theta_2 = 1$.

1. Estimate the probability that a student who studies for 40 h and has an undergrad GPA of 3.5 gets an A in the class.
2. How many hours would the student in part (a) need to study to have a 50\% chance of getting an A in the class?

--------

Solution

--------

1. Based on the coefficients, the model is
    $$ P(Y = 1 | x_1 = 40, x_2 = 3.5) = \frac{1}{1 + e^{-(-6 + (0.05)(40) + 3.5)}} = \frac{1}{1 + e^{0.5}} \approx 0.38
    $$
    so the student has a 38\% chance of getting an A.
2. To have a 50\% chance of getting an A, we solve for $x$ in:
    $$P(Y = 1 | x_1 = x, x_2 = 3.5) = .5 \Leftrightarrow
    \frac{1}{1 + e^{-(-6 + 0.05 x + 3.5)}} = \frac{1}{2}\Rightarrow 1 = e^{-(-2.5 + 0.05 x)} \Rightarrow x = 50.
    $$
    so the student would have to study for 50 hours.  

```python
predictors = ['gre', 'gpa']

x = dfAdmissions[predictors]
y = dfAdmissions['admit']

admitLogRegModel = LogisticRegression(penalty='l2', solver = 'newton-cg').fit(x, y)

# theta0 is our intercept
admittheta0 = admitLogRegModel.intercept_[0]

# theta consists of all other thetas
admittheta1 = admitLogRegModel.coef_[0][0]
admittheta2 = admitLogRegModel.coef_[0][1]

# Print out thetas
print('The thetas for the Admissions Dataset are:\ntheta0 =', admittheta0)
print('theta1 =', admittheta1)
print('theta2 =', admittheta2)
```
